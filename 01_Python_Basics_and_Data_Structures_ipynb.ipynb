```python
#Task: List Comprehensions - Phase 2 - LINQ Challenge
#Task: Create a list of numbers from 1 to 20. Using range

nums = list(range(1,21)) #21 is exclusive. It will not be part of the result. range(start, stop, step) .
# Pythonic way to filter and transform data - Append doesn't return anything. Modifies the existing list.
even_squares = list(x**2 for x in nums if x%2==0) ''' If you already have a generator in parentheses (), you wrap it in the list() function.
                                                      This is technically what .ToList() does in C#.'''
even_squares_1 = [x**2 for x in nums if x%2==0] # [] acts as the ToList() signal to the compiler

print(f"even_squares = {even_squares}")
print(f"even_squares_1 = {even_squares_1}")
```
