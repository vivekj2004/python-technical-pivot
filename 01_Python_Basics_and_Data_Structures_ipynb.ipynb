```python
#Task: List Comprehensions - Phase 2 - LINQ Challenge
#Task: Create a list of numbers from 1 to 20. Using range

nums = list(range(1,21)) #21 is exclusive. It will not be part of the result. range(start, stop, step) .
# Pythonic way to filter and transform data - Append doesn't return anything. Modifies the existing list.
even_squares = list(x**2 for x in nums if x%2==0) ''' If you already have a generator in parentheses (), you wrap it in the list() function.
                                                      This is technically what .ToList() does in C#.'''
even_squares_1 = [x**2 for x in nums if x%2==0] # [] acts as the ToList() signal to the compiler

print(f"even_squares = {even_squares}")
print(f"even_squares_1 = {even_squares_1}")
```

```python
# Goal: Prove that a for loop is just a wrapper for iter() and next().
my_tech = ['Cloud', 'Data', 'AI']
# Get the Pointer:
my_iter = iter(my_tech)
#Manually pull values:
print(next(my_iter))
print(next(my_iter))
print(next(my_iter))
# - print(next(my_iter)) - This will throw StopIteration exception in Python.
''' "Python uses exceptions to signal the end of a collection, whereas C# uses a boolean return from MoveNext()."   '''
```

```python
# Phase 4: Forensic Inspection (The dir() Method)
# Goal: See the hidden interface (dunder methods) that makes an object iterable
print (dir(my_tech))
print (dir(45)) # - this does not have __iter__
''' dir() reveals why some things are "Iterable" and others aren't.
If the "Magic Method" __iter__ exists, Python treats it like it implements IEnumerable. '''
```
