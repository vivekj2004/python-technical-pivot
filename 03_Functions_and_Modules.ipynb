```python
#Block 1: The formal function
def find_max(data: list[int]) -> int:
  max_item = data[0]
  for item in data:
    if(item>max_item):
      max_item = item
  return max_item

print(find_max([-1, -5, 0, 4, 10, 5]))
```

```python
#Block 2: The Scope Test
count = 0
def increment():
  global count #Try removing this line to see the 'UnboundLocalError'
  count += 1
  return count

print(increment())
```

```python
#Block 3: Generator function
''' A generator function that returns factors '''
def factors(n:int):
  for k in range(1, n+1):
    if n%k == 0:
      yield k

print(list(factors(21)))
```
