```python
# Graceful exit - Exception Handling
data = ['Cloud', 'Data']
it = iter(data)
try:
    print(next(it))
    print(next(it))
    print(next(it)) # This will trigger the error
except StopIteration:
    print("End of list reached successfully.")
finally:
    print("System cleanup complete.")
```
